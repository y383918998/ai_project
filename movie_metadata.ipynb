```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from wordcloud import WordCloud
```

```python
filepath = './dataset/movie_metadata.csv/movie_metadata.csv'
```

```python
df = pd.read_csv(filepath)
df
```

```python
plt.figure(figsize=(15, 15))
sns.histplot(data=df, x='imdb_score', kde=True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('IMDB Score', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.savefig('./figure/movie_metadata/IMDB_Score.png', dpi=600)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
sns.histplot(data=df, x='title_year', kde=True)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.savefig('./figure/movie_metadata/Year.png', dpi=400)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
sns.jointplot(data=df, x='imdb_score', y='duration')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('IMDB_Score', fontsize=15)
plt.ylabel('Duration', fontsize=15)
plt.savefig('./figure/movie_metadata/IMDB_Score_Duration.png', dpi=400)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.title('Correlation of Movie Features\n', fontsize=18, weight=600)
sns.heatmap(df.corr(), annot=True)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig('./figure/movie_metadata/heatmap.png', dpi=400, bbox_inches='tight')
plt.show()
```
